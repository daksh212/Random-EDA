Checking if the test dataset for duplicates

```python
# Check for duplicates, this adds a new column to the dataset
test_df["is_duplicate"]= df.duplicated()

print(f"#total= {len(test_df)}")
print(f"#duplicated= {len(test_df[test_df['is_duplicate']==True])}")
```

Checking if the train dataset for duplicates

```python
# Check for duplicates, this adds a new column to the dataset 
train_df["is_duplicate"]= df.duplicated()

print(f"#total= {len(train_df)}")
print(f"#duplicated= {len(train_df[train_df['is_duplicate']==True])}")
```

Dropping the columns in both since there are no duplicates in either

```python
# Drop the duplicate rows using index - best way to drop in pandas
index_to_drop = test_df[test_df['is_duplicate']==True].index
test_df.drop(index_to_drop, inplace=True)

# Remove the duplicate marker column
test_df.drop(columns='is_duplicate', inplace=True)
print(f'#total= {len(test_df)}')
```

```python
# Drop the duplicate rows using index - best way to drop in pandas
index_to_drop = train_df[train_df['is_duplicate']==True].index
train_df.drop(index_to_drop, inplace=True)

# Remove the duplicate marker column
train_df.drop(columns='is_duplicate', inplace=True)
print(f'#total= {len(train_df)}')
```

```python
#Seeing NaN in train
display(train_df[train_df['Age'].isnull()])
display(train_df[train_df['Embarked'].isnull()])
display(train_df[train_df['Cabin'].isnull()])
```

For test, Cabin and age should be dropped and for train, cabin and age should be dropped too. For these columns, the number of NaNs is too much, and hence imputing these large number of values would skew the data and result in inaccurate predictions

```python
train_df.drop(columns='Age', inplace=True)
train_df.drop(columns='Cabin', inplace=True)
test_df.drop(columns='Age', inplace=True)
test_df.drop(columns='Cabin', inplace=True)
```

```python
#imputing the NaNs in the other columns. Fairly clear why we would use mean for fare but mode for embarked
means = {c:test_df[c].mean() for c in test_df.columns if test_df[c].dtype != object}
mode = {c:train_df[c].mode() for c in train_df.columns if train_df[c].dtype == object}

print(f"mean-fare= {means['Fare']}")
print(f"mode-embarked= {mode['Embarked']}")


test_df['Fare'] = test_df['Fare'].fillna(means['Fare'])
```

```python
train_df['Embarked'] = train_df['Embarked'].fillna('S')
```

I will be dropping name since PassengerID is a  primary key, and name might have first or last name duplicates

```python
train_df.drop(columns='Name', inplace=True)
test_df.drop(columns='Name', inplace=True)
train_df.drop(columns='Ticket', inplace=True)
test_df.drop(columns='Ticket', inplace=True)
```

```python
#time to one hot encode, similar thing can be done with pd.get_dummies but I saw this code function online. 
# It helped me understand the workings behind one hot encoding

def encode_onehot(_df, _f):
    _df2 = pd.get_dummies(_df[_f], prefix='', prefix_sep='').groupby(level=0, axis=1).max().add_prefix(_f+' - ')
    _df3 = pd.concat([_df, _df2], axis=1)
    _df3 = _df3.drop([_f], axis=1)
    return _df3




# Print nominal variables
for f in list(train_df.columns.values):
    if train_df[f].dtype == object:
        print(f) 

for f in list(test_df.columns.values):
    if test_df[f].dtype == object:
        print(f) 

#train_df= encode_onehot(train_df, 'Name')
train_df= encode_onehot(train_df, 'Sex')
#train_df= encode_onehot(train_df, 'Ticket')
train_df= encode_onehot(train_df, 'Embarked')

#test_df= encode_onehot(test_df, 'Name')
test_df= encode_onehot(test_df, 'Sex')
#test_df= encode_onehot(test_df, 'Ticket')
test_df= encode_onehot(test_df, 'Embarked')
```

```python
cols = []
for f in list(test_df.columns.values):
    if 'Embarked' in f:
        cols += [f]
for f in list(test_df.columns.values):
    if 'Sex' in f:
        cols += [f]  


for f in list(train_df.columns.values):
    if 'Embarked' in f:
        cols += [f]
for f in list(train_df.columns.values):
    if 'Sex' in f:
        cols += [f]  
# Displaying the onehot-encoded        
display(test_df[cols][:10])
```
